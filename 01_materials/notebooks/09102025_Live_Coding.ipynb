```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
from sklearn import set_config

# Output dataframes instead of arrays
set_config(transform_output="pandas")
```

```python
sacramento = pd.read_csv("dataset/sacramento.csv")
sacramento
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Linear regression to find line of best fit between house square footage and house cost.  
# y = b0 + b1x.  house_price = b0 + b1(house_size)

# Calculate beta coefficients to minimize distance between predicted and real data points.  least squares.
```

```python
# Output dataframes instead of arrays
set_config(transform_output="pandas")
```

```python
# load in our data.

sacramento = pd.read_csv('dataset/sacramento.csv')
sacramento
```

```python
sacramento_train, sacramento_test = train_test_split(sacramento, train_size = 0.75, random_state=42)
```

```python
# Step 1: Define our x and y

X_train = sacramento_train[['sq__ft']]
y_train = sacramento_train['price']
```

```python
# Step 2: Call and initialize linear regression model.

lm = LinearRegression()
```

```python
# Step 3: fit our model to our data

lm.fit(X_train, y_train)
```

```python
lm.coef_

#beta1, b1 value.  slope of line.  Increase in ~$139 per square foot.
```

```python
lm.intercept_
```

```python
# Put beta values into a pandas data frame

pd.DataFrame({
    'slope': [lm.coef_],
    'intercept': [lm.intercept_]
})
```

```python
# house price is 7069 + 139(house_size)
```

```python
# Evaluate model on unseen test data.  Evaluate model performance.

sacramento_test['predicted'] = lm.predict(sacramento_test[['sq__ft']])


rmpse = mean_squared_error(
    y_true = sacramento_test['price'],
    y_pred =  sacramento_test['predicted']
)**0.5
rmpse
```

```python
sacramento_test
```

```python
r2 = r2_score(
    y_true= sacramento_test['price'],
    y_pred= sacramento_test['predicted']
)
r2
```

```python
# This line calculates the minimum and maximum values of the "sq__ft" column in the sacramento dataset.
sqft_prediction_grid = sacramento[["sq__ft"]].agg(["min", "max"])

# Uses the linear model to predict prices for the min and max square footage values.
sqft_prediction_grid["predicted"] = lm.predict(sqft_prediction_grid)

# Plot the original data
plt.scatter(sacramento["sq__ft"], sacramento["price"], label='Original Data')

# Plot the model predictions as a line
plt.plot(sqft_prediction_grid["sq__ft"], sqft_prediction_grid["predicted"], color='red', label='Model Predictions')

# Add labels and legend
plt.xlabel('House size (square feet)')
plt.ylabel('Price (USD)')
plt.title('Scatter Plot of House Size vs Price')
plt.legend()

# Show the plot
plt.show()
```

```python
returned_dictionary = cross_validate(
    estimator=lm,
    cv=5,    # setting up the cross validation number
    X= sacramento[["sq__ft"]],
    y= sacramento["price"],
    scoring="neg_root_mean_squared_error" #or scoring="r2"
)

```

```python
cv_5_df = pd.DataFrame(returned_dictionary)    # Converting it to pandas DataFrame

cv_5_df
```

```python
cv_5_df["test_score"] = cv_5_df["test_score"].abs()
cv_5_df
```

```python
cv_5_metrics = cv_5_df.agg(["mean","sem"])
cv_5_metrics
```

```python
# Multivariable linear regression to include more than one predictor

# initialize a multivariable linear regression

mlm = LinearRegression()
```

```python
mlm.fit(
    sacramento_train[['sq__ft', 'beds']],
    sacramento_train['price']
)
```

```python
mlm.coef_

# first number is effect of square footage, second is number of bedrooms (holding house size constant).
```

```python
mlm.intercept_
```

```python
#make predictions on our test data
sacramento_test['predicted'] = mlm.predict(sacramento_test[['sq__ft', 'beds']])
```

```python
mlm_rmspe = mean_squared_error(
    y_true = sacramento_test['price'],
    y_pred = sacramento_test['predicted']
)**0.5
mlm_rmspe
```

```python
mlm_r2 = r2_score(
    y_true = sacramento_test['price'],
    y_pred = sacramento_test['predicted']
)
mlm_r2
```

```python
# Five fold cross validation to assess performance.

returned_dictionary_mlm = cross_validate(
    estimator=mlm,
    cv = 5, 
    X = sacramento[['sq__ft', 'beds']],
    y = sacramento['price'],
    scoring = 'r2'
)
```

```python
cv_5_df_mlm = pd.DataFrame(returned_dictionary_mlm)
cv_5_df_mlm
```

```python
cv_5_df_mlm.agg(['mean', 'sem'])
```
