### Classification using KNN

Let's set up our workspace and use the **Wine dataset** from `scikit-learn`. This dataset contains 178 wine samples with 13 chemical features, used to classify wines into different classes based on their origin.

#### **Question 2:** 
#### Standardization and data-splitting

Next, we must preform 'pre-processing' or 'data munging', to prepare our data for classification/prediction. For KNN, there are three essential steps. A first essential step is to 'standardize' the predictor variables. We can achieve this using the scaler method, provided as follows:

```python
# Select predictors (excluding the last column)
predictors = wine_df.iloc[:, :-1]

# Standardize the predictors
scaler = StandardScaler()
predictors_standardized = pd.DataFrame(scaler.fit_transform(predictors), columns=predictors.columns)

# Display the head of the standardized predictors
print(predictors_standardized.head())
```

(i) Why is it important to standardize the predictor variables?

> Your answer here...

The predictor variables are all measured in different units and with different scales.  If they are not standardized, one parameter with a large scale may disproportionately influence classification.

(ii) Why did we elect not to standard our response variable `Class`?

> Your answer here...

The variable class is not a predictor and it is not directly measured.  This variable is assigned based on (standardized) data from all the other variables.

#### **Question 3:**
#### Model initialization and cross-validation
We are finally set to fit the KNN model. 


Perform a grid search to tune the `n_neighbors` hyperparameter using 10-fold cross-validation. Follow these steps:

1. Initialize the KNN classifier using `KNeighborsClassifier()`.
2. Define a parameter grid for `n_neighbors` ranging from 1 to 50.
3. Implement a grid search using `GridSearchCV` with 10-fold cross-validation to find the optimal number of neighbors.
4. After fitting the model on the training data, identify and return the best value for `n_neighbors` based on the grid search results.

```python
# Your code here...
# 1. Initializing knn

knn = KNeighborsClassifier(n_neighbors=5)
knn
```

```python
# 2. Defining a parameter grid.
parameter_grid = {
    "n_neighbors": range(1, 51)
}
```

```python
# 3. Implementing a grid search

wine_tune_grid = GridSearchCV(
    estimator=knn,
    param_grid=parameter_grid,
    cv=10
)
```

```python
# 4. Fitting the model on the training data

# If I were adding the 'y' values back in, here's what I'd do.

# wine_tune_grid.fit(
#    wine_train.drop('class', axis=1),
#    wine_train['class']
#)

# Since I incorporated 'y' during the train_test_split, here's what I'll do:

wine_tune_grid.fit(
    wine_train,
    class_train
)
```

```python
# 4. Returning best value for k based on fitting results

wine_tune_grid.best_params_
```

```python
# It seems the output from best_params is a dictionary.  I want to learn more.  
wine_tune_grid.best_params_.keys()
```

```python

ideal_k = wine_tune_grid.best_params_['n_neighbors']
print(f'The best value for "k" is {ideal_k}.')
```

#### **Question 4:**
#### Model evaluation

Using the best value for `n_neighbors`, fit a KNN model on the training data and evaluate its performance on the test set using `accuracy_score`.

```python
# Your code here...

# Fit the model on the training data using n_neighbors ideal_k value.


# If I had waited until after the train_test_split to define the 'y' variable, here's the code:

# knn = KNeighborsClassifier(n_neighbors=ideal_k)
# X = wine_train.drop('class', axis=1)
# y = wine_train['class']

# Since I defined both 'X' and 'y' during train_test_split, here's the code I'll use:

knn = KNeighborsClassifier(n_neighbors=ideal_k)
X = wine_train
y = class_train

knn.fit(X, y)
```

```python
# Predict on the test set

wine_test['predicted'] = knn.predict(wine_test)
```

```python
len(wine_test['predicted'])
```

```python
# compute the score on validation data

y_test = class_test
y_pred = wine_test['predicted']
wine_score = accuracy_score(y_test, y_pred)
print(f' The accuracy score is {wine_score}')
```
